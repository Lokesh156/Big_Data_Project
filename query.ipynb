```python
reviews=spark.read.format('csv').option('header','true').load('Result.csv')
reviews.createOrReplaceTempView('hotel_reviews')
reviews.show()
```

```python

average_scores_over_time = reviews.groupBy("Review_Date").avg("Reviewer_Score")
average_scores_over_time.show()
```

```python

from pyspark.sql.functions import col

reviews = reviews.withColumn("Reviewer_Score", col("Reviewer_Score").cast("double"))

average_scores_over_time = reviews.groupBy("Review_Date").avg("Reviewer_Score")
average_scores_over_time.show()
```

```python
print(average_scores_over_time.count())
```

```python
average_scores_over_time.write.csv("goal_01.csv",header=True,quoteAll=True)
```

```python
top_hotels = reviews.groupBy("Hotel_Name").avg("Reviewer_Score").orderBy("avg(Reviewer_Score)", ascending=False).limit(25)
top_hotels.show(25)
```

```python
top_hotels.write.csv("goal_02.csv", header=True)
```

```python
reviewer_scores_distribution = reviews.select("Reviewer_Score")
reviewer_scores_distribution.show()
```

```python
reviewer_scores_distribution.write.csv("goal_03.csv",header=True)
```

```python
from pyspark.sql.functions import explode, split, col

tags_df = reviews.select("Tags")
tags_df = tags_df.withColumn("Tag", explode(split(col("Tags"), ",")))
most_common_tags = tags_df.groupBy("Tag").count().orderBy("count", ascending=False)
most_common_tags.show()
print(most_common_tags.count())
```

```python
most_common_tags.write.csv("goal_04",header=True)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

positive_reviews = reviews.filter(col("Positive_Review") != "No Positive").select("Positive_Review")
positive_text = " ".join(positive_reviews.rdd.flatMap(lambda x: x).collect())

wordcloud = WordCloud().generate(positive_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from pyspark.sql.functions import year, month

temporal_review_counts = reviews.withColumn("Year", year("Review_Date")).withColumn("Month", month("Review_Date"))
temporal_review_counts = temporal_review_counts.groupBy("Year", "Month").count().orderBy("Year", "Month")
temporal_review_counts.show()
```

```python
from pyspark.sql.functions import udf
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from pyspark.sql.types import DoubleType

# Initialize SentimentIntensityAnalyzer from NLTK
sia = SentimentIntensityAnalyzer()

# Define a UDF for sentiment analysis
def analyze_sentiment(text):
    sentiment_score = sia.polarity_scores(text)
    return sentiment_score['compound']

# Register the UDF
sentiment_udf = udf(analyze_sentiment, DoubleType())

# Apply sentiment analysis to both positive and negative reviews
hotel_reviews_df = reviews.withColumn("Positive_Sentiment", sentiment_udf("Positive_Review"))
hotel_reviews_df = reviews.withColumn("Negative_Sentiment", sentiment_udf("Negative_Review"))

# Show the results
hotel_reviews_df.select("Positive_Review", "Positive_Sentiment", "Negative_Review", "Negative_Sentiment").show()
```
